```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
iris_data = pd.read_csv('IRIS.csv')
```

```python
iris_data.head()
```

```python
iris_data.shape #no. of rows and cols.
```

```python
iris_data.describe()
```

```python
iris_data['species'].value_counts() #Counts should be similar values
```

```python
iris_data.groupby('species').mean() # mean of each column for each species
```

```python
#seperating features and labels
X= iris_data.drop(columns='species', axis=1)
Y= iris_data['species']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training -> Logistic Regression

```python
model= LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train) #compare prediction with original labeled data
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test) #compare prediction with original labeled data
print(testing_data_accuracy)
```

Predicting species

```python
input_data= (6.4,3.2,5.3,2.3)

# Convert input data into numpy array for faster processing
input_data_numpy = np.asarray(input_data)

#reshape the np array as we are predicting for one instance
input_data_reshaped= input_data_numpy.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)
```
